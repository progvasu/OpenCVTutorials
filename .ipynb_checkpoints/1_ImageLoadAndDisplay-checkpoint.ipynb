```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# audio and video are interchangeable
```

```python
img = cv2.imread('butter.jpeg', cv2.IMREAD_GRAYSCALE)
# cv2.IMREAD_GRAYSCALE = 0
# default 'IMREAD_COLOR' without alpha = 1
# IMREAD_UNCHANGED = -1
```

```python
# showing image in a window
cv2.imshow('title - image', img)
cv2.waitKey(0) # wait for any key to be pressed
cv2.destroyAllWindows()
```

```python
# showing image using matplotlib
# we cant show color because openCV BGR and matplotlib is RGB
plt.imshow(img) # colors are messed up
```

```python
# instead we need to do
plt.imshow(img, cmap='gray')
```

```python
plt.imshow(img, cmap='gray', interpolation='bicubic') # smooths out the image?
```

```python
# can plot on this image
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.plot([300, 100], [50, 300], 'c', linewidth=5)
```

```python
# saving an image
cv2.imwrite('gray_butter.jpeg', img)
```
